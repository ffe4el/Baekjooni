```python
#2748 피보나치 수 2

# 탑다운 방식을 사용해서 풀어보자 => 재귀함수 사용
import sys

# input = sys.stdin.readline

def f(n):
    if n == 0 :
        return 0
    if n == 1 :
        return 1
    return f(n-1)+f(n-2)

print(f(int(input())))
```

```python
#2748 피보나치 수 2
# 탑다운, 메모이제이션(캐시) 활용!
import sys

# input = sys.stdin.readline

cache = [-1]*91 #0~90까지
cache[0]=0
cache[1]=1

def f(n):
    if cache[n]==-1:
        cache[n]=f(n-1)+f(n-2)
    return cache[n]

print(f(int(input())))
```

```python
#2748 피보나치 수 2
#바텀업 방식으로 풀기 => for문 사용

n = int(input())
cache=[-1]*91
cache[0]=0
cache[1]=1

for i in range(2,n+1):
    cache[i] = cache[i-1] + cache[i-2]

print(cache[n])
# 이방식은 그냥 0부터 n까지 피보나치수열을 캐시에 다 저장해놓는 방법이다
```

```python
#11051 이항계수2
# 재귀로... 메모이제이션 사용X
import sys


mod = 10007
sys.setrecursionlimit(10**7)
n = int(input())
r = int(input())

def bino(n,r):
    if r==0 or r==n:
        return 1
    return bino(n-1,r-1)+bino(n-1,r)

print(bino(n,r))
```

```python
#11051 이항계수2
# 재귀로... 메모이제이션 사용 O
import sys


mod = 10007 #나누는수
sys.setrecursionlimit(10**7)
n,r = map(int,input().split())
cache=[[0]*1001 for i in range(1001)] #2차원배열 만들기

def bino(n,r):
    if cache[n][r] !=0:
        return cache[n][r]
    if r==0 or r==n:
        return 1
    else : #cache[n][r]가 0이면
        cache[n][r] = bino(n-1,r-1) + bino(n-1,r)
        cache[n][r] %= mod
    return cache[n][r]

print(bino(n,r))
```

```python
#11051 이항계수2
#for문 바텀업 방식으로도 풀어보자..!

mod = 10007
n,r = map(int,input().split())
cache = [[0]*1001 for i in range(1001)]

for i in range(1001):
    cache[i][0] = cache[i][i] = 1
    for j in range(i):
        cache[i][j] = cache[i-1][j-1] + cache[i-1][j]
        cache[i][j] %= mod

# for i in range(7):
    # print(cache[i])

# print()
print(cache[n][r])
```

```python
# 11726 2*n 타일링
# 음너무 어려운데ㅎㅎ 탑다운방법으로 해볼까...? => 개같이 실패
# 이거... 약간 넘사적 접근인데...ㅠㅠ
# 결론 f(n) = f(n-1)+f(n-2) 로 계산하는거래 시봉뱅
# 물론 모두 이러진 않고,,, 어쩌다 점화식이 이렇게 나와버린것임..

n = int(input())
cache = [0]*1001

def f(n):
    if cache[n] != 0:
        return cache[n]
    elif n==1:
        return 1
    elif n==2:
        return 2
    else :
        cache[n] = f(n-1)+f(n-2)
        cache[n] %= 10007
    return cache[n]

print(f(n))



```

```python
# 11726 2*n 타일링
# 이번엔 동일문제 바텀업으로 풀어보기..!

n = int(input())
cache = [0]*1001
cache[1]=1
cache[2]=2

for i in range(3,1001):  #O(N)이란다...?
    cache[i] = (cache[i-1]+cache[i-2])%10007

print(cache[n])




```

```python
# 10844 쉬운 계단 수

# 바텀업으로 먼저....!

mod = 1000000000
# cache[n][d] =길이가 n, 마지막숫자가 d인 계단의 개수
cache = [[0]*10 for _ in range(101)]

for j in range(1,10):
    cache[1][j]=1

for i in range(2,101):
    for j in range(10):
        if j>0:
            cache[i][j] += cache[i-1][j-1]
            cache[i][j] %= mod
        if j<9:
            cache[i][j] += cache[i-1][j+1]
            cache[i][j] %= mod

ans = 0
n = int(input()) #자릿수 입력
for j in range(10):
    if cache[n][j]:
        ans += cache[n][j]
        ans %= mod

print(ans)
```
