```python
#2748 피보나치 수 2

# 탑다운 방식을 사용해서 풀어보자 => 재귀함수 사용
import sys

# input = sys.stdin.readline

def f(n):
    if n == 0 :
        return 0
    if n == 1 :
        return 1
    return f(n-1)+f(n-2)

print(f(int(input())))
```

위에 코드는 아직 캐싱을 적용하지 않았기 때문에 입력숫자가 커질때 시간초과가 생기기마련이다...
아래 코드에서 메모이제이션을 활용해보자!

```python
#2748 피보나치 수 2
# 탑다운, 메모이제이션(캐시) 활용!
import sys

# input = sys.stdin.readline

cache = [-1]*91 #0~90까지
cache[0]=0
cache[1]=1

def f(n):
    if cache[n]==-1:
        cache[n]=f(n-1)+f(n-2)
    return cache[n]

print(f(int(input())))
```

```python
#2748 피보나치 수 2
#바텀업 방식으로 풀기 => for문 사용

n = int(input())
cache=[-1]*91
cache[0]=0
cache[1]=1

for i in range(2,n+1):
    cache[i] = cache[i-1] + cache[i-2]

print(cache[n])
# 이방식은 그냥 0부터 n까지 피보나치수열을 캐시에 다 저장해놓는 방법이다
```
