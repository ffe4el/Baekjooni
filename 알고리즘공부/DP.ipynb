```python
#11051 2
# ...   O
import sys


mod = 10007 #
sys.setrecursionlimit(10**7)
n,r = map(int,input().split())
cache=[[0]*1001 for i in range(1001)] #2 

def bino(n,r):
    if cache[n][r] !=0:
        return cache[n][r]
    if r==0 or r==n:
        return 1
    else : #cache[n][r] 0
        cache[n][r] = bino(n-1,r-1) + bino(n-1,r)
        cache[n][r] %= mod
    return cache[n][r]

print(bino(n,r))
```

```python
# 10844   

#  ....!

mod = 1000000000
# cache[n][d] = n,  d  
cache = [[0]*10 for _ in range(101)]

for j in range(1,10):
    cache[1][j]=1

for i in range(2,101):
    for j in range(10):
        if j>0:
            cache[i][j] += cache[i-1][j-1]
            cache[i][j] %= mod
        if j<9:
            cache[i][j] += cache[i-1][j+1]
            cache[i][j] %= mod

ans = 0
n = int(input()) # 
for j in range(10):
    if cache[n][j]:
        ans += cache[n][j]
        ans %= mod

print(ans)
```
